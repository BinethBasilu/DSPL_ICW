```python
# delete first row
df_p = df_p.drop(df_p.index[0])
df_p
```

```python
#change data types
df_p['date'] = pd.to_datetime(df_p['date'])
df_p['price'] = df_p['price'].astype(float)
df_p['latitude'] = df_p['latitude'].astype(float)
df_p['longitude'] = df_p['longitude'].astype(float)
df_p['usdprice'] = df_p['usdprice'].astype(float)
df_p.info()
```

```python
# to take the summary statistic
df_p.describe()
```

```python
#rename admin1 as province and admin 2 as district
df_p = df_p.rename(columns={'admin1': 'Province', 'admin2': 'District'})
df_p
```

```python

df_p['priceflag'].value_counts()
```

```python
#delete priceflag column
if 'priceflag' in df_p.columns:
    #delete priceflag column
    df_p = df_p.drop(['priceflag'], axis=1)
    print("Column 'priceflag' deleted successfully.")
else:
    print("Column 'priceflag' not found in the DataFrame.")

df_p
```

```python
#replace missing values of Province and District replace with National Average
df_p['Province'].fillna('National Average', inplace=True)
df_p['District'].fillna('National Average', inplace=True)
```

```python
#replace missing values of latitude and longtitude with 0
df_p['latitude'].fillna(0, inplace=True)
df_p['longitude'].fillna(0, inplace=True)
```

```python
#check the count of the records
df_p.count()
```

```python
df_p['market'].value_counts()
```

```python
df_p['market'] = df_p['market'].replace(['Mulaitivu'], 'Mullativu')
df_p['market'] = df_p['market'].replace(['Keppetipola (DEC}'], 'Keppetipola')
df_p['market'] = df_p['market'].replace(['Meegoda(DEC)'], 'Meegoda')


df_p['market'].value_counts()
```

```python
import matplotlib.pyplot as plt
economic_centers = ['Economic Centre-Dambulla', 'Economic Centre - Peliyagoda', 'Economic Centre-Pettah', 'Fish market-Peliyagoda', 'Fish market-Negombo', 'Economic Centre-Maradagahamula']
filtered_df = df_p[df_p['market'].isin(economic_centers)]
commodity_counts = filtered_df.groupby(['market', 'category'])['category'].count().unstack()
commodity_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Categories by Economic Centers')
plt.xlabel('Economic Center')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='cathgory')
plt.tight_layout()
plt.show()
```
